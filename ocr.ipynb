# OCR Nedir?

**OCR (Optical Character Recognition)** — Türkçesiyle **Optik Karakter Tanıma**, taranmış belge, fotoğraf veya ekran görüntüsü gibi görsellerdeki yazılı içerikleri dijital ve düzenlenebilir metne dönüştüren bir teknolojidir.



## Nasıl Çalışır?

1. **Görüntü Girişi:** Belge veya yazı içeren bir görüntü alınır.
2. **Ön İşleme:** Görüntü netleştirilir, arka plan temizlenir, yazı alanları ayrıştırılır.
3. **Karakter Tanıma:** Görseldeki harf ve sayılar analiz edilerek metne dönüştürülür.
4. **Metin Çıktısı:** Tanınan içerik, kopyalanabilir ve düzenlenebilir dijital metin olarak çıkar.



## Kullanım Alanları

| Alan                   | Açıklama                                                      |
|------------------------|---------------------------------------------------------------|
|  Eğitim ve Arşivleme | Kitapların, belgelerin dijital arşivlenmesi                  |
|  Bankacılık          | Kimlik, fatura, çek gibi belgelerden bilgi okuma              |
|  Mobil Uygulamalar   | Kamera ile belge tarayıp yazıyı metne çevirme (örnek: CamScanner) |
|  Veri Girişi         | Formlardaki bilgilerin otomatik olarak sisteme girilmesi       |
|  Arama Sistemleri    | Görsellerdeki yazının indekslenip aranabilir hale getirilmesi  |



## Popüler OCR Araçları

## Sınırlamaları

```python
from ultralytics import YOLO
```

```python
model = YOLO("yolov8n.pt")
```

```python
model.train(data="data.yaml", epochs=1, imgsz=640, batch=8)
```

```python
import pytesseract
import cv2 as cv
model2 = YOLO("runs/detect/train/weights/best.pt")
```

```python
img_predict = cv.imread("27.jpg")
```

```python
results = model2.predict(source=img_predict, conf=0.5)
```

```python
for box in results[0].boxes:
    x1, y1, x2, y2 = box.xyxy[0].int().tolist()
    plate_img = img_predict[y1:y2, x1:x2]
    gray_plate = cv.cvtColor(plate_img, cv.COLOR_BGR2GRAY)
    custom_config = r'--oem 3 --psm 7'
    plate_text = pytesseract.image_to_string(gray_plate, config=custom_config, lang='eng')
    print("Plaka:", plate_text)
```
