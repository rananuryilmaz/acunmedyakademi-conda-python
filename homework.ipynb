### Pandas kütüphanesinde 10 adet fonksiyon araştırılıp kullanım örneği olan bir ipynb dosyası ile paylaşılacak.

```python
import pandas as pd
```

```python
df4 = pd.read_csv("Flight_Price_Dataset_of_Bangladesh.csv")
df4
```

```python
#1.
#İstatistiksel özet almak için kullanılan fonksiyon:
# Bunlara,sayısal sütunlardaki değerlerin minimum, maksimum, ortalama (merkezi eğilim), standart sapma (dağılım) ve verilerdeki tüm boş olmayan değerlerin sayısı dahildir.
df4.describe() 
```

```python
#2.
#Bu fonksiyon DataFrame'in hızlı bir özetini döndürür. Bu, aşağıda gösterildiği gibi sütun adları ve ilgili veri türleri,
#eksik değerler ve DataFrame tarafından kullanılan bellek hakkında bilgi içerir.
df4.info()
```

```python
#3.
#Veri kümesindeki eksik (NaN) değerleri tespit etmek için kullanılır.
df4.isnull().sum()
```

```python
#4.
#Bu fonksiyon havayolu ismine göre alfabetik sıralama yapar.Sıralama fonksiyonu diyebiliriz.
df4.sort_values("Airline")
```

```python
#5.
#sıralamyı tersten yapmak istersek bu fonksiyonu kullanırız.
df4.sort_values("Airline", ascending=False)
```

```python
# 6.
# Bu fonksiyon bir değerin bir sütunda kaç kez göründüğünü döndürür. 
df4.value_counts("Class")
```

```python
# 7.
# verileri havayolu firmalarına göre grupluyoruz ve her bir havayolunun ortalama uçuş fiyatını hesaplıyoruz.
# kısaca havayolu firmasının ortalama bilet fiyatını öğrenmek için GROUPBY ile gruplandırdık MEAN ile ortalamasını aldık.
df4.groupby("Airline")["Base Fare (BDT)"].mean()
```

```python
# 8.
#Bir sütunda benzersiz hangi değerler olduğunu gösterir.
df4["Airline"].unique()
```

```python
# 9.
# rename fonksiyonu sütun adlarını değiştirmeye yarar. base fare sütunun adını price olarak değiştirdik.
# inplace ile kalıcı hale getirdik
df4.rename(columns={"Base Fare (BDT)":"Price"}, inplace=True)
```

```python
#Burada görüldüğü üzere price olarak değiştirmiş.
print(df4.columns)
```

```python
#10.
#veri kümesindeki en küçük birkaç değeri elde etmenize yardımcı olur.
df4.nsmallest(7, "Days Before Departure")
```
